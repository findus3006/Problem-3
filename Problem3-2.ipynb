```python
import numpy as np
import cmath as cm
import matplotlib.pyplot as plt

def euler(t,a,b):
    dt = t[2] - t[1]; #stepsize
    r = np.zeros([2,np.size(t)]); #Position
    p = np.zeros([2,np.size(t)]); #Momentum
    r[:,0] = a; #initial conditions
    p[:,0] = b;
    for i in range(np.size(t)-1):
        r[:,i+1] = r[:,i] + dt*1*p[:,i];
        p[:,i+1] = p[:,i] + dt*(-1/(np.linalg.norm(r[:,i])**3))*r[:,i];
    return p,r

px = 1/plt.rcParams['figure.dpi'] #Define pixel for plotting later
p0 = np.array([0,0.3,1,2]) # Initial value of momentum in y direction
dt = np.array([0.5,0.1,0.01,0.001]); #stepsize

#Stepsize is varied
fig, ax = plt.subplots(4,4,figsize=(2560*px, 1440*px))
for k in range(np.size(dt)):
    t = np.arange(0,100,dt[k]); #Time
    #Damping constant is varied
    for j in range(np.size(p0)):
        p,r = euler(t,[1, 0],[0, p0[j]]); #Euler method
        
        # Plotting
        ax[j][k].grid(True),ax[j][k],ax[j][k].set_title('$\Delta$t = '+ str(dt[k]) + ' and ' + '$p_0$ = '+ str(p0[j]))
        ax[j][k].plot(r[0,:],r[1,:],label='Position Euler')
        ax[j][k].plot(p[0,:],p[1,:],label='Momentum Euler')
        ax[j][k].legend()
        #plt.close();
        
```
